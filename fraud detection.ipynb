# Feature Selection

```python
#library
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import cross_val_score
```

```python
#prepare the threshold data
threshold=np.linspace(0,rf2.feature_importances_.max(),20)
```

```python
#to find the best threshold for feature selection
score=[]
for i in threshold:
    #do the feature selection when threshold=i and use the rf2 model
    x_embedded=SelectFromModel(rf2,threshold=i).fit_transform(x_resampled,y_resampled)
    #calculate the cross validation score
    once=cross_val_score(rf2,x_embedded,y_resampled,cv=5).mean()
    score.append(once)
```

```python
#plot the score to find the best feature selection
plt.plot(threshold,score)
plt.show()
```

```python
#see the result after feature selection
x_embedded=SelectFromModel(rf2,threshold=0.058).fit_transform(x_resampled,y_resampled)
score=cross_val_score(rf2,x_embedded,y_resampled,cv=5).mean()
score
```
